**Series** = a one dimensional array that can hold any data type

```python
data = [100, 102, 104]
series = pd.Series(data, index = ['a', 'b', 'c'])
print(series)
```

.Series is a constructor for a series object, it has some parameters like the array, and index. probably more exist

how can we access data from a series object? using .loc, also known as location.

```python
print(series.loc['a'])
print(series.loc['b'])
print(series.loc['c'])
```

you can also use loc to modify values. same thing as like an array.

```python
series.loc['b'] = 499
print(series)
```

we can also access values by integer location using iloc. same thing.

```python
print(series.iloc[2])
```

lets filter values.

```python
new_data = [100,102, 104,200, 202]
new_series = pd.Series(new_data, index = ['a', 'b', 'c', 'd', 'e'])
print(new_series[new_series >= 200])
```

more complex filtering example.


```python
calories = {"Day1":1750 , "Day2":2100, "Day3": 1700}
pizza_series = pd.Series(calories)
pizza_series.loc["Day3"] += 2100 #what a fatass
print(pizza_series[pizza_series > 2000])
```

**Dataframes**

A dataframe is tabular data structure with rows and collumns. it is similar to what excel is.

```python
data = {"Name": ["Spongebob", "Patrick", "Squidward"],
        "Age": [30, 35,50],
        }
df = pd.DataFrame(data, index = ["Employee 1", "Employee 2", "Employee 3"])
print(df.loc["Employee 2"])
print(df.iloc[2])
```

How can we add a new collumn or row?

```python
df["Job"] = ["Cook", "NA", "Cashiner"]
print(df)

new_rows = pd.DataFrame([
                        {"Name" :"Sandy", "Age": 28, "Job":"Engineer"},
                        {"Name" :"Krabs", "Age": 60, "Job":"Manager"}]
                       ,index = ["Employee 4","Employee 5"])
df = pd.concat([df ,new_rows ])
print(df)
```
