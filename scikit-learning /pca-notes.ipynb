pca - prinipal component analysis
a way to reduce dimensionality
basically trying to find the features that contribute to the variance the most - eliminating the features that are somwhat similar

```python
from sklearn.datasets import fetch_openml
from sklearn.decomposition import PCA
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split


x, y = fetch_openml('mnist_784', return_X_y=True, as_frame=False)
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=0)
```

pca does reduce the amount of information available, but you can decide how many components you want to get rid off and compare performance

```python
pca = PCA(n_components=20)
x_train_reduced = pca.fit_transform(x_train)
x_test_reduced = pca.transform(x_test)

x_train.shape
x_train_reduced.shape
```

```python
from sklearn.preprocessing import StandardScaler
clf = LogisticRegression(max_iter=100)
scaler = StandardScaler()
x_train_reduced_scaled = scaler.fit_transform(x_train_reduced)
x_test_reduced_scaled = scaler.transform(x_test_reduced)
clf.fit(x_train_reduced_scaled, y_train)
print(clf.score(x_test_reduced_scaled, y_test))
```

```python
x_train_scaled = scaler.fit_transform(x_train)
x_test_scaled = scaler.transform(x_test)
clf.fit(x_train_scaled, y_train)
print(clf.score(x_test_scaled, y_test))
```
