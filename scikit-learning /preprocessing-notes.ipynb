```python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

data = load_iris()
X = data.data
y = data.target

# it is important to scale all data, as when using some sort of distance based model, for example k neighbours, one of the features might be way more senstive if it much greater than another feature. scaling equalized all features

scaler = StandardScaler()

x_train, x_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

x_train_scaled = scaler.fit_transform((x_train)) #when wanting to learn what scale to use, we add fit to the function.
x_test_scaled = scaler.transform((x_test)) # when wanting to just scale, we only use the transform method.
x_train_scaled
```

```python
# CAN ALSO DO IT MANUALLY
import numpy as np
(x_train - np.mean(x_train, axis=0)) / np.std(x_train, axis=0)
```

```python
#Min max scaling
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()

x_train_scaled = scaler.fit_transform((x_train)) #when wanting to learn what scale to use, we add fit to the function.
x_test_scaled = scaler.transform((x_test))
x_train_scaled
```

```python
#manual min max
x_min = np.min(x_train, axis=0)
x_max = np.max(x_train, axis=0)
(x_train - x_min) / (x_max - x_min)
```
