```python
import sklearn

from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
data = load_iris()
x = data.data
y = data.target
```

splitting the data set into train and test is important, just like how mentioned in 240. the train test split function allows us to easily split the data into whatever proportions that we want.


```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2)
#the four variables are important. they are what the train test split function actually spits out.
```

```python
import numpy as np
import matplotlib.pyplot as plt

counts = np.bincount(y_train)
positions = np.arange(3)

plt.bar(positions, counts)
plt.xticks(positions, data.target_names)
```

There is also issues with splitting the data like this. can have a disproportionate split, have training be on mainly one category while the test data contains the opposite.

```python
from sklearn.model_selection import StratifiedShuffleSplit #this guarantees that train and test split are even
split = StratifiedShuffleSplit(n_splits=1, test_size=0.2)

for train_index, test_index in split.split(x, y):
    x_train, x_test = x_train[train_index], x_test[test_index]
    y_train, y_test = y_train[train_index], y_test[test_index]
```
