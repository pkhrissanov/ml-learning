```python
import numpy as np
from sklearn.datasets import fetch_openml
from sklearn.preprocessing import OrdinalEncoder

data = fetch_openml('car', as_frame=True).frame
data
```

some data sets are easier to do feature encoding with than others. in this case, stuff like lug_boot, and safety can be very easily put onto a scale. in other examples, it could be harder as something like countries dont have a natural scale. will be first doing ordinal (on a scale) encoding.

```python
import numpy as np
colums_to_enconde = np.array(['lug_boot', 'safety'])


encoder = OrdinalEncoder(
    categories= [
                ['small', 'med', 'big'],
                ['low', 'med', 'high']]
)

data[colums_to_enconde] = encoder.fit_transform(data[colums_to_enconde])
data
```

```python
encoder.inverse_transform(data[colums_to_enconde])
data
```

This covers how to encode scaling categories, now onto more sophisticated ones, we will be learning the one hot encoder?

```python
data = fetch_openml('adult', as_frame=True).frame
data
```

```python
from sklearn.preprocessing import OneHotEncoder
data.occupation.value_counts()
import pandas as pd
encoder = OneHotEncoder(handle_unknown='ignore', sparse_output=False) #handle unknown decides the behaviour of when we get something we have not seen before.
encoded_values = encoder.fit_transform(data[['occupation', 'race']])
new_collumns = encoder.get_feature_names_out(['occupation', 'race'])
encoded_values
new_collumns

df_encoded = pd.DataFrame(encoded_values, columns=new_collumns, index = data.index)
data_final = pd.concat([
    df_encoded, data.drop(columns = ['occupation', 'race']
    )], axis = 1)
data_final

```

# Summary
This chapter goes over how to encode features of the data set that are not numerical. There are two ways that were covered. First, for features that have a natural scaling, we can use OrdinaryEncoder to give each feature a level.
For non natural scaling, we can instead use One hot encoding which basically just gives you the binary represenation of what level the feature is.
