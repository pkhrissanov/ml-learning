```python
from jupyter_client.provisioning.provisioner_base import KernelProvisionerMeta
from sklearn.datasets import make_blobs
from sklearn.cluster import k_means, KMeans
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
```

```python
x, _ = make_blobs(n_samples= 5000, centers= 5, random_state= 10)
scaler = StandardScaler()

x_scaled = scaler.fit_transform(x)
```

```python
plt.scatter(x_scaled[:, 0], x_scaled[:, 1])
```

```python
kmeans = KMeans(5)
kmeans.fit(x_scaled)
plt.scatter(x_scaled[:, 0], x_scaled[:, 1], c = kmeans.labels_)
```

```python
from sklearn.datasets import make_moons
x, _ = make_moons(n_samples= 5000, noise = 0.02, random_state= 10)
scaler = StandardScaler()
x_scaled = scaler.fit_transform(x)

kmeans = KMeans(2)
kmeans.fit(x_scaled)
plt.scatter(x_scaled[:, 0], x_scaled[:, 1], c = kmeans.labels_) # this is a issue with k means, as it only works with euclidiean distance. there are things that can solve this=, for example db means
```

```python
from sklearn.cluster import DBSCAN
dbscan = DBSCAN(eps = 0.2)
dbscan.fit(x_scaled)
plt.scatter(x_scaled[:, 0], x_scaled[:, 1], c = dbscan.labels_)
```
